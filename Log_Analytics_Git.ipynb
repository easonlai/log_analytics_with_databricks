```
# Spark read log json file.
df = spark.read.json("wasbs://"+container_name+"@"+storage_account_name+".blob.core.windows.net/y=2021/m=06/d=27/h=10/m=00/*.json")
df.printSchema()
df.show()
```

```
# Display Spark dataframe from log json file.
display(df)
```

```
# Select LogEntry column from Spark dataframe.
df.select("LogEntry").show(truncate=False)
```

```
# Import Panda library
import pandas as pd
```

```
# Convert Spark dataframe to Pandas dataframe.
df_panda = df.select("LogEntry").toPandas()
```

```
# Display Pandas dataframe.
display(df_panda.head())
```

```
# Extract/Flatten log data inside LogEntry column.
df_panda_a = df_panda.LogEntry.str.split(",", expand=True,)
# Rename columns header to meaningful name.
df_panda_a.columns = ['timestamp','server_ip','process_id','log_type','class_name','client_ip','message']
df_panda_a.head()
```

```
# Check Pandas dataframe info.
df_panda_a.info()
```

```
# Convert timestamp column to datetime data type.
df_panda_a['timestamp']= pd.to_datetime(df_panda_a['timestamp'])
df_panda_a.head()
```

```
# Check Pandas dataframe info.
df_panda_a.info()
```

```
# Covert curated Pandas dataframe back to Spark dataframe.
df_2 = spark.createDataFrame(df_panda_a)
```

```
# Display Pandas dataframe.
display(df_2)
```

```
# Display type of logs in log_type.
display(df_2.groupBy("log_type").count())
```

```
# Making new Spark dataframe which filtered TRACE only.
df_3 = df_2.filter(df_2.log_type.like("%TRACE%"))
display(df_3)
```

```
# Display TRACE dataframe and group by message and client_ip.
display(df_3.groupBy("message","client_ip").count())
```

```
# Making new Spark dataframe which filtered ERROR only.
df_3err = df_2.filter(df_2.log_type.like("%ERROR%"))
display(df_3err)
```

```
# Display ERROR dataframe and group by message and client_ip.
display(df_3err.groupBy("message","client_ip").count())
```
